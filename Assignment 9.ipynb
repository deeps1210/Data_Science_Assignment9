```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

```python
#Solution 1 : Delete unnamed columns
```

```python
df.head()
```

```python
#first finding out all the unnamed columns and then dropping it
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1).head()
```

```python
#Solution 2 : Show the distribution of male and female
```

```python
df.pivot_table(values = 'Id', index = ['Gender'], columns = 'Year', aggfunc = 'count')
```

```python
#Solution 3: Top 5 most preferred names
```

```python
df_top_5 = df.groupby(['Name'])['Id'].agg('count').sort_values(ascending = False).head(5)
```

```python
df_top_5
```

```python
#Solution 4 :What is the median name occurence in the dataset
```

```python
df[df.Id == df.median()['Id']]
```

```python
#Solution 5: Show the distribution of male and female count by states
```

```python
df.pivot_table(values = 'Count', index = ['Gender', 'Year'], columns = 'State', aggfunc = 'count')
```
